# 15) For each quantitative attribute, calculate its average, standard deviation, minimum, mode, range and maximum values.

```python
l1 = ['age','balance','duration','campaign']
for i in l1:
    print(i)
    
    print("mean",data[i].mean())
    print("standard deviation",data[i].std())
    print("minimum",data[i].min())
    print("maximum",data[i].max())
    print("mode",data[i].mode()[0])
    print("range",data[i].max() - data[i].min())
    
    print("--------------------")
    
```
